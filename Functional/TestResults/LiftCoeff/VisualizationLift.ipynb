```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Abbot = np.loadtxt("./AbbotRe6mil.txt")
Circular = np.loadtxt("./CLCircularP1.txt")
Elliptical = np.loadtxt("./CLEllipticalP2.txt")
```

```python
fig, ax = plt.subplots()
ax.plot(Abbot[:,0], Abbot[:,1], label='Abbot, Re = 6mil', color='blue', linestyle='--')
ax.plot(Circular[:,0], Circular[:,1], label='Circular, P = 1', color='red', linestyle='-')
ax.plot(Elliptical[:,0], Elliptical[:,1], label='Elliptical, P = 2', color='green', linestyle='-')

ax.set_xlabel('Angle of Attack (degrees)')
ax.set_ylabel('Lift Coefficient')
ax.legend()
plt.show()
```

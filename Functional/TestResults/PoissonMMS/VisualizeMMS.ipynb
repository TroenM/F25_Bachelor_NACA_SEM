```python
import matplotlib.pyplot as plt
import firedrake as fd
import numpy as np
```

```python
errors = np.loadtxt("MMS_Poisson_Errors.txt")
dof = np.loadtxt("MMS_Poisson_DOF.txt")
```

```python
for i in range(3):
    plt.loglog(dof[i,:], errors[i,:], label=f"P={i+1}", marker='o')

plt.xlabel("Degrees of Freedom")
plt.ylabel("L2 Error")
plt.title("MMS on Poisson Equation, with phi = sin(x)*sin(y)")
plt.legend()
```
